
# Q4. Explain the distinctions between indexing and slicing.

“Indexing” means referring to an element of an iterable by its position within the iterable.

“Slicing” means getting a subset of elements from an iterable based on their indices.

# Q5. What happens if one of the slicing expression&#39;s indexes is out of range?

Indexing raises an out-of-range error when trying to access out of bound elements in a sequence. 

But slicing doesn’t raise an out-of-range error. Slicing handles out-of-range indices so you won’t run into IndexError with slicing.


# Q6. If you pass a list to a function, and if you want the function to be able to change the values of the list—so that the list is different after the function returns—what action should you avoid?

```python
var = 10
lst = [1, 2, 3]

def func():
    var = 11
    lst[:] = [1, 2, 3, 4]
    return var, lst

var1, lst1 = func()
print(var, lst) 
print(var1, lst1)
```


# Q7. What is the concept of an unbalanced matrix?

A matrix is balanced if all cells in the matrix are balanced and a cell of the matrix is balanced if the number of cells in that matrix that are adjacent to that cell is strictly greater than the value written in this cell. 
Adjacent cell means cells in the top, down, left, and right cell of each cell if it exists. 

<b>Input</b>: N = 3, M = 3                                                                                                     
mat[][] = {{1, 2, 3}, 
{4, 5, 6}, 
{7, 8, 9}}                                                                                                                    
<b>Output</b>: Unbalanced                                                                                                     
<b>Explanation</b>: Each cell of the given grid is not stable, so the overall grid is unbalanced.

<b>Input</b>: N = 3, M = 3                                                                                                     
mat[][] = {{1, 2, 1}, 
{2, 3, 2}, 
{1, 2, 1}}                                                                                                                      
<b>Output</b>: Balanced                                                                                                        
<b>Explanation</b>: Each cell of the given grid is stable, so the overall grid is Balanced. 


# Q8. Why is it necessary to use either list comprehension or a loop to create arbitrarily large matrices?

Because we need to create arrays until N (number of rows) and M (number of columns) defined is reached.

```python
c_size = int(input("Enter size of column: "))
r_size = int(input("Enter size of row: "))
x = []
y = []
for j in range(0, c_size):
    y.append(0)
for i in range(0, r_size):
    x.append(y)
print(x)
```
